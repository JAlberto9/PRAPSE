```python
import pandas as pd
import requests
from tabulate import tabulate
from bs4 import BeautifulSoup
from tabulate import tabulate



query = {'anio': '2018','mes': '8','delegacion': '14','distrito': '-1','municipio': '-1','moda': '3','ciclo': '3','tipoprograma': '0','nivel': '3','consultar': 'si'}
req = requests.get('http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenDelegacion.do?', params=query)
soup = BeautifulSoup(req.text, "lxml")
#Borrar todos los valors que contengan subtotaltes 
for div in soup.find_all("div", {'class':'SubTotal'}): 
    div.decompose() #  Borra los las filas con la calse subtotal


tabla=soup.find_all('tr')
report_tables=soup.find_all('table',class_='table table-responsive table-striped table-bordered')


    
```

```python
res = []
for tr in tabla:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)


df = pd.DataFrame(res, columns=["Distrito", "Municipio", "Producto", "Sembrada","Cosechada","Obtenida(ton)","Obtenido()"])
print(df)
```
